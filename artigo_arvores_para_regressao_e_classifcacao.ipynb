```python
#CÓDIGO ÁRVORE - CLASSIFICAÇÃO

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
# Criando o conjunto de dados para Decisão de Aprovação de Empréstimo
data_emprestimo = {
    'Renda': [50000, 60000, 45000, 50000, 75000, 35000, 40000, 120000, 65000,
                    47000, 58000, 35000],
    'Solicitado': [15000, 30000, 12000, 40000, 35000, 20000, 10000, 50000, 38000,
                            14000, 36000, 17000],
    'Aprovacao': ['Aprovado', 'Aprovado', 'Aprovado', 'Negado', 'Aprovado', 'Negado',
                  'Aprovado', 'Negado', 'Negado', 'Aprovado', 'Negado', 'Negado']
}

# Convertendo para DataFrame
df_emprestimo = pd.DataFrame(data_emprestimo)

# Mapeando a variável categórica 'Aprovacao' para numérica
df_emprestimo['Aprovacao'] = df_emprestimo['Aprovacao'].map({'Aprovado': 1, 'Negado': 0})

# Definindo as variáveis independentes (X) e a dependente (y)
X_emprestimo = df_emprestimo[['Renda', 'Solicitado']]
y_emprestimo = df_emprestimo['Aprovacao']

# Criando o modelo de Árvore de Decisão
clf_emprestimo = DecisionTreeClassifier(random_state=45)
clf_emprestimo = clf_emprestimo.fit(X_emprestimo, y_emprestimo)

# Plotando a árvore de decisão
plt.figure(figsize=(20,20))
tree.plot_tree(clf_emprestimo, feature_names=['Renda', 'Solicitado'],
               class_names=['Negado', 'Aprovado'], filled=True)
plt.show()
```

```python
#CÓDIGO ÁRVORE - REGRESSÃO

import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import matplotlib.pyplot as plt

# Criando o conjunto de dados
data = {
    'Preco': [ 18.8, 20.7, 22.99, 36.99, 39.99, 41.2, 50.5, 68.2, 80.4],
    'Quantidade': [108, 72, 94, 10, 13, 18, 31, 35, 49]
}

# Convertendo para DataFrame
df = pd.DataFrame(data)

# Separando as variáveis independentes (X) e dependentes (y)
X = df[['Preco']]
y = df['Quantidade']

# Criando o modelo de Árvore de Regressão com profundidade máxima de 2
regressor = DecisionTreeRegressor(random_state=45, min_samples_leaf=3)
regressor.fit(X, y)

# Plotando a árvore de regressão
plt.figure(figsize=(12,8))
tree.plot_tree(regressor, feature_names=['Preco'], filled=True)
plt.show()
```
